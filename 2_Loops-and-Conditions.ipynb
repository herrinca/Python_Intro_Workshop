## **Operators**

### Comparison Operators
In programming, we often need to compare values to make decisions or determine relationships between them. For this purpose, we use comparison operators that allow us to compare two values and evaluate conditions such as greater than, less than, or equal to. 



All camparison operators return a bool, True or False 

````python
a = 'A'
b = 'C'

print("a > b", a > b)    
print("a >= b", a >= b)  
print("a < b", a < b)    
print("a < b", b < a)    
print("a <= b", b <= a)  
print("a == b", a == b)  
print("a != b", a != b) # Not equal is probably the only operator you have not seen before 
````

<br>

### Logical Operators
Unlike other programming languages python uses keywords and, or and not for logical operators which can be used to combine conditional statements:



````python
# and = True is both are True
print('True and True: ', True and True)
````

````python
# or = True if at least one is True
print('True or False: ', True or False)
print('False or False:', False or True)
print('False or False:', False or False)
````

````python
# not changes bool
print('not True  :', not True)      
print('not False :', not False)     
````

````python
# feel free to play around with value and type changes
a = 1
b = 3
c = 5

print('a > b and c > a:', a > b and c > a) 
print('a > b and c < a:', a > b and c < a) 
print('a < b and c < a:', a < b and c < a) 

print('a > b or c > a:', a > b or c > a) 
print('a > b or c < a:', a > b or c < a) 
print('a < b or c < a:', a < b or c < a) 

print('not a > b:', not a > b)     
````

## **Python Condition Statements**
In python and other programming languages the key word if is used to check if a condition is true and to execute the block code. **Remember the indentation after the colon.**

<br>

**Python Indentation**

Indentation refers to the spaces at the beginning of a code line. Where in other programming languages the indentation in code is for readability only, the indentation in Python is very important.

Python uses indentation to indicate a block of code.
### *if* statement
```# syntax
if condition:
    this part of code runs for truthy conditions
```

````python
# one condition example
a = 3
if a > 0:
    print('A is a positive number')
````

As you can see in the example above, 3 is greater than 0. The condition was true and the block code was executed. However, if the condition is false, we do not see the result. In order to see the result of the falsy condition, we should have another block, which is going to be else.

### *if else* statement

If condition is true the first block will be executed, if not the else condition will run.

```
# syntax
if condition:
    this part of code runs for truthy conditions
else:
     this part of code runs for false conditions
```

````python
# two condition example
a = 3
if a < 0:
    print('A is a negative number')
else:
    print('A is a positive number')
````

### *if elif else* statement

Often in programming we need to test more than two conditions. We can use elif when we have multiple conditions.

```
# syntax
if condition:
    code
elif condition:
    code
else:
    code
```

````python
# multi-conditional example
a = 0
if a > 0:
    print('A is a positive number')
elif a < 0:
    print('A is a negative number')
else:
    print('A is zero')
````

## **Loops**
Life is full of routines. In programming we also do lots of repetitive tasks. In order to handle repetitive task programming languages use loops. 

### *while* loops
We use the reserved word while to make a while loop. It is used to execute a block of statements repeatedly until a given condition is satisfied. When the condition becomes false.

```
# syntax
while condition:
    code goes here
```

````python
# conditional loop example
count = 0
while count < 4:
    print(count)
    count = count + 1

# code after loop will excute after condition becomes False
print( f"while condition failed at {count}")
````

### *break* within a loop
We use *break* when we like to get out of or stop the loop.


```
# syntax
while condition:
    code goes here
    if another_condition:
        break
```

````python
# break example
count = 0
while count < 5:
    print(count)
    count = count + 1
    if count == 3:
        break
````

### *for* loops
A *for* loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string). 

```
# syntax
for iterator in lst:
    code goes here
```

*for* loop with a list

````python
# for loop with list example
numbers = [0, 1, 2, 3, 4, 5]
for num in numbers: # num is temporary name to refer to the list's items, valid only inside this loop
    print(num)       # the numbers will be printed line by line, from 0 to 5
````

*for* loop with a string

````python
# for loop with string example
language = 'Python'
for letter in language:
    print(letter)
````

*for* loop with a dictionary

````python
person = { 'first_name':'Asabeneh',
           'last_name':'Yetayeh',
           'age':250,
           'country':'Finland'}
````

````python
# with labels or keys
for key in person.keys():
    print(key)
````

````python
# with values
for key in person.values():
    print(key)
````

````python
# with both keys and values
for key, value in person.items():
    print(key, value) # this way we get both keys and values printed out
````

### *continue* within a loop
We use continue when we like to skip to the next iterator in the loop. *break* ends the loop while *continue* just skips to the next item. 

```
# syntax
for iterator in sequence:
    code goes here
    if condition:
        continue
```


````python
numbers = (0,1,2,3,4)
for number in numbers:
    print(number)
    if number == 3:
        continue
    print('Next number should be ', number + 1)
print('outside the loop')
````

### *range()* function

To loop through a set of code a specified number of times, we can use the *range()* function,
The *range()* function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```
# syntax
for iterator in range(start, end, step):
```

````python
# loop through 6 iterations
for x in range(6):
    print(x)
````

### Nested loops
We can write loops inside a loop.

```
# syntax
for x in y:
    for t in x:
        print(t)
```

````python
adj = ["red", "tasty", "yellow"]
fruits = ["apple", "cherry", "banana"]

for x in adj:
    for y in fruits:
        print(x, y)
````

### *zip()* with *for* loops
*zip()* is builtin function that is used to combine two or more iterables into a single iterable.

````python
adj = ["red", "tasty", "yellow"]
fruits = ["apple", "cherry", "banana"]

for x, y in zip(adj,fruits):
    print(x, y)
````

### *enumerate()* with *for* loops
enumerate() lets you write Pythonic for loops when you need a count and the value from an iterable

````python
fruits = ["apple", "banana", "orange", "grape"]

for index, fruit in enumerate(fruits):
    print(f"Index: {index}, Fruit: {fruit}")
````
